```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_loss_acc(df):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(df['step'], df['train_cost'], label='Train Loss', color='blue')
    plt.plot(df['step'], df['test_cost'], label='Test Loss', color='orange')
    plt.xlabel('Step')
    plt.ylabel('Loss')
    plt.title('Train/Test Loss')
    plt.legend()

    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(df['step'], df['train_acc'], label='Train Accuracy', color='green')
    plt.plot(df['step'], df['test_acc'], label='Test Accuracy', color='red')
    plt.xlabel('Step')
    plt.ylabel('Accuracy')
    plt.title('Train/Test Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

### 0.05M_1k

```python
df=pd.read_csv('1k_0.05M.csv')
plot_loss_acc(df)
```

### 0.05M_10k

```python
df=pd.read_csv('10k_0.05M.csv')
plot_loss_acc(df)
```
